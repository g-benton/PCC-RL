```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fpath = "./saved-outputs/"

together_rwrds = torch.load(fpath + "together_saved_rwrds.pt")
together_rates = torch.load(fpath + "together_saved_rates.pt")

sep_rwrds = torch.load(fpath + "saved_rwrds.pt")
sep_rates = torch.load(fpath + "saved_rates.pt")

iterations = torch.arange(sep_rwrds.shape[1])
```

```python
mean_sep = sep_rwrds.mean(0)
mean_tog = together_rwrds.mean(0)
```

```python
ax_fs = 18
lwd = 3

colors = sns.color_palette()
sns.set_style("white")
plt.figure(figsize=(10, 6))

plt.plot(iterations, mean_sep[:, 0], label="Separate", color=colors[0],
        linewidth=lwd)
plt.plot(iterations, mean_sep[:, 1], color=colors[0], linewidth=lwd)

plt.plot(iterations, mean_tog[:, 0], label="Together", color=colors[1],
         linewidth=lwd)
plt.plot(iterations, mean_tog[:, 1], color=colors[1], linewidth=lwd)
        
plt.legend(fontsize=ax_fs)
plt.xlabel("Iteration", fontsize=ax_fs)
plt.ylabel("Reward", fontsize=ax_fs)
plt.tick_params("both", labelsize=ax_fs-2)
sns.despine()
plt.title("Multiple Senders",  fontsize=ax_fs+4)
plt.savefig("./multiple_senders.pdf", bbox_inches='tight')
plt.show()
```

```python
trial = 2
colors = sns.color_palette()
sns.set_style("white")
plt.figure(figsize=(10, 6))

plt.plot(iterations, sep_rates[trial, :, 0], label="Separate Sender ", color=colors[0],
        linewidth=lwd, marker='.', linestyle="None")
# plt.plot(iterations, sep_rates[trial, :, 1], color=colors[0], linewidth=lwd)

plt.plot(iterations, together_rates[trial, :, 0], label="Together Sender ", color=colors[1],
         linewidth=lwd, marker='.', linestyle="None")
# plt.plot(iterations, together_rates[trial, :, 1], color=colors[1], linewidth=lwd)
        
plt.legend(fontsize=ax_fs)
sns.despine()
plt.show()
```
