```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load in PCC Data

```python
pcc_dat = np.load("./run_model_out.npz")
```

```python
pcc_dat.files
```

```python
pcc_actions = pcc_dat['actions']
pcc_rwrds = pcc_dat['pcc_rwrds']
```

```python
pcc_actions
```

## Load in BOCC Data

```python
bocc_actions = torch.load("../acquisition-testing/actions.pt")
bocc_rwrds = torch.load("../acquisition-testing/rwrds.pt")
```

## Plot Actions

```python
sns.set_style("white")

fs = 18
lwd = 2.5
ind = 13


plt.figure(figsize=(10, 6))
plt.plot(bocc_rwrds[1, ind, :], linewidth=lwd,
        label="BOCC")
plt.plot(pcc_rwrds[:100], linewidth=lwd,
        label="PCC-RL")

## add cubic here:

plt.xlabel("Iteration", fontsize=fs)
plt.ylabel("Reward", fontsize=fs)
plt.tick_params("both", labelsize=fs-2)
plt.title("Example Simulation", fontsize=fs+2)

## IF YOU ADD ANOTHER CURVE YOU'LL NEED TO CHANGE bbox_to_anchor
plt.legend(fontsize=fs-2, bbox_to_anchor=(0.72, -0.15), ncol=3)
sns.despine()
plt.savefig("./compare_pcc_bocc.pdf", bbox_inches='tight')
```

```python
sns.set_style("white")

fs = 18
lwd = 2.5
ind = 13


plt.figure(figsize=(10, 6))
plt.plot(bocc_actions[1, ind, :], linewidth=lwd,
        label="BOCC")
plt.plot(pcc_actions[:100, 1], linewidth=lwd,
        label="PCC-RL")

## add cubic here:

plt.xlabel("Iteration", fontsize=fs)
plt.ylabel("Action", fontsize=fs)
plt.tick_params("both", labelsize=fs-2)
plt.title("Example Simulation", fontsize=fs+2)

## IF YOU ADD ANOTHER CURVE YOU'LL NEED TO CHANGE bbox_to_anchor
plt.legend(fontsize=fs-2, bbox_to_anchor=(0.72, -0.15), ncol=3)
sns.despine()
plt.savefig("./pcc_bocc_action.png", bbox_inches='tight')
```
